## 2. 다음중 출력되는 코드 행은??

```javascript
function abc() {
    var a = 'bbb';
    console.log(a);
    
    function c() {
        console.log(a);
        
        (function() {
            console.log(a);
            a = 'ccc';
        })();
        
        var a;
        console.log(a);
    }
    
    function d() {
        console.log(a);
    }
    c();
    a = 'ddd';
    d();
};
abc();
```

## 6. console.log 에 찍히는 값을 적으시오.

```javascript
function Child(aa, bb) {
    if(this.constructor !== Child) new Child();
    this.aa = aa;
    this.bb = bb;
}

function Parent() {
    if(this.constructor !== Parent) new Parent();
}

Parent.prototype.a = [1];
Parent.prototype.f = function() {
    this.a.push(this.a.length + 1);
}

Child.prototype = new Parent();
Child.prototype.f = function() {
    this.a.push(this.a.length + 1);
}

Child.prototype.constructor = Child
var p = new Parent();
var c = new Child();

p.f();
c.f();

console.log(p.a);
console.log(c.a);
```

## 7. a, b, c, d 의 값을 구하시오

```javascript
function closure(start) {
    return function count() {
        return {
            inc: function() { start++ },
            get: function() { return start; }
        }
    }
}

var f1 = closure(0);
var f2 = f1;
var f3 = closure(0);

var fc1 = f1();
var fc2 = f2();
var fc3 = f3();
var fc4 = fc1;

fc1.inc();
fc2.inc();
fc3.inc();
fc4.inc();

var a = fc1.get();
var b = fc2.get();
var c = fc3.get();
var d = fc4.get();

console.log(a);
console.log(b);
console.log(c);
console.log(d);
```
